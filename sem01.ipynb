Потом, нужен, конечно, класс для музыкального инструмента. Попробуйте реализовать музыкальные инструменты с помощью наследования: можно создать некий абстрактный класс "инструмент", а от него будут наследовать классы "гитара", "пианино" и так далее - при этом у гитар и пианино могут быть свои разные характеристики, н-р, у гитары количество струн, тип порожка и так далее. Можно и более подробную иерархию, если будете успевать (например, инструменты бывают электрическими и акустическими). Наконец, можно создать класс "покупатель", у которого будет кошелек. 

Наконец, неплохо бы не забыть написать некий код, который будет приводить наши классы в движение: имитировать рабочий день магазина (можно в цикле), с рандомной вероятностью посещение покупателей, давать нашему владельцу-пользователю возможность решить, не пора ли закупить новые инструменты или, может быть, затеять рекламную кампанию. 

Фантазия всячески поощряется, могу за нее поставить дополнительные баллы.

```python
class Instrument():
   def __init__(self, name):
      self.name = name
      self.cost_price = 10
      

      if self.name.lower().find('electr') >= 0:
          self.cost_price *= 1.3
          
   
   def count_price(self):
      return round(self.cost_price + self.cost_price * 0.3, 2)


class StringInstrument(Instrument):
    def __init__(self, name, strings : int):
        super().__init__(name) 
        self.strings = strings
        self.cost_price = self.cost_price * 1.6 + self.strings * 0.7

    def __repr__(self):
        return f"{self.__class__.__name__}('{self.name}')"
   



class WindInstrument(Instrument):
    def __init__(self, name, weight : float):
        super().__init__(name) 
        
        if weight < 0:
            self.weight = -1 * weight
        elif weight > 10:
            self.weight = 10
        else:
            self.weight = weight

        self.cost_price =  self.cost_price * 1.4 + self.weight * 0.7

    def __repr__(self):
         return f"{self.__class__.__name__}('{self.name}')"
    
    

class PercussionInstrument(Instrument):
    def __init__(self, name, weight : float):
        super().__init__(name) 

        if weight < 0:
            self.weight = -1 * weight
        elif weight > 100:
            self.weight = 100
        elif weight < 10:
            self.weight = 10
        else:
            self.weight = weight
        
        self.cost_price =  self.cost_price * 1.7 + self.weight * 0.1

    def __repr__(self):
         return f"{self.__class__.__name__}('{self.name}')"
    
```

```python
class Seller():
    def __init__(self, name, professionalism : int):
        self.name = name
        if professionalism < 1:
            self.professionalism = 1
        elif professionalism > 10:
            self.professionalism = 10
        else: 
            self.professionalism = professionalism
        self.mood = randint(-10, 10)  
        self.salary = randint(30, 50) * self.professionalism
        self.rate = 0

    

    def serve_customer(self, shop, item, desired_quantity):
        if 1 <= self.professionalism <= 3:
            service_quality = self.professionalism + (self.mood / 4 if self.mood >= 0 else self.mood)
        elif 4 <= self.professionalism <= 7:
            service_quality = self.professionalism + (self.mood / 2)
        else:
            service_quality = self.professionalism + (self.mood / 4 if self.mood <= 0 else self.mood)
    
        if shop.warehouse[item]['amount'] <= 0:
            print(f"Customer wanted to buy {desired_quantity} {item}, but it's out of stock.")
            shop.reputation -= 2 
            self.mood -= 2
        elif 0 < shop.warehouse[item]['amount'] < desired_quantity:
            if service_quality >= randint(0, 5): 
                print(f"Customer wanted to buy {desired_quantity} {item}, but only {shop.warehouse[item]['amount']} were available. Fortunately, {self.name} did a great job!")
                self.rate += 1
                self.mood += 1
            else:
                print(f"Customer wanted to buy {desired_quantity} {item}, but only {shop.warehouse[item]['amount']} were available. Unfortunately, {self.name} upset them even more")
                shop.reputation -= 1
                self.rate -= 1
                self.mood -= 3
            shop.cash += shop.warehouse[item]['price'] * shop.warehouse[item]['amount']
            shop.warehouse[item]['amount'] -= shop.warehouse[item]['amount']
        else:
            if service_quality >= randint(0, 10):  
                print(f"Customer bought {desired_quantity} {item} from {self.name}. They are happy!")
                shop.reputation += 3
                self.rate += 1
                self.mood += 3
            else:
                print(f"Customer bought {desired_quantity} {item} from {self.name}, but they are unhappy.")
                shop.reputation -= 1 
                self.rate -= 1 
                self.mood -= 2
            shop.cash += shop.warehouse[item]['price'] * desired_quantity
            shop.warehouse[item]['amount'] -= desired_quantity
            

    
```

```python
class Game:
    def __init__(self, shop_name):
        self.shop = Shop(shop_name)

    def count_customers(self):
        num_customers = randint(int(self.shop.reputation / 10), int(self.shop.reputation / 2))
        return num_customers

    def working_day(self):
        for x in range(self.count_customers()):
            item = choice(list(self.shop.warehouse.keys())) 
            desired_quantity = randint(1, 3) 
            seller = choice(self.shop.stuff)
            seller.serve_customer(self.shop, item, desired_quantity)
            sleep(randint(1,7))
        print(f"End of the day.\nReputation: {self.shop.reputation}\nCash: {self.shop.cash}")
    
    def start_of_the_game(self):
        if input('Do you want to start a new game? (y/n)') == 'n':
            pass
        print(f"You're starting a new Music Shop '{self.shop.name}'. Reputation: {self.shop.reputation}, Cash: {self.shop.cash}")
        print('You need to order instruments to sell')
        self.shop.order_supply()
        print("Warehouse contents:")
        self.shop.check_the_warehouse()
        self.shop.hire_stuff()
        print(f"Great! Here is your stuff:")
        self.shop.show_stuff()

        print(f"You reputation now is {self.shop.reputation}. It effects the amount of customers that comes to you. You can try and run a pr compaign but remember that it can both increase and decrease your reputation.")
        if input('Do you want to run a pr compaign now? y/n') == "y":
            self.shop.pr_compaign()
        else:
            pass

        
    def end_of_day(self):
        self.shop.pay_salaries()
        print("Warehouse contents:")
        self.shop.check_the_warehouse()
     
    
        if input("Do you want to  order supplies (y/n)?") == 'y':
            self.shop.order_supply()
        else:
            pass

        print("Here ir your sellers' rating. You can raise or decrease salaries for some of them, it will change their mood and professionalism.")
        for x in self.shop.stuff:
            print(f"{x.name}\t{x.rate}")

        if input("Do you want to change someone's salary (y/n)?") == 'y':
            seller = self.shop.find_seller_by_name(input("Whose salary do you want to change (Write the seller's name)?"))
            if seller:
                action = input(f"Do you want to raise or decrease the salary? (raise/decrease): ").lower()
                percent = float(input("Enter the percentage to change the salary (number only)"))
                if action == 'raise':
                    self.shop.raise_salary(seller, percent)
                elif action == 'decrease':
                    self.shop.decrease_salary(seller, percent)
            else:
                print(f"You wrote a wrong name.")
        else:
            pass


    def between_days(self):
        while True:
            act = input("Choose an action: order new instruments (o), check the warehouse (w), run a pr compaign (p), check the stuff (c), change worker's salary(s), hire new workers (h), delete an instrument from good's list (d). If you're ready to start a new day write 'e'")
            if act == 'o':
                self.shop.order_supply()
            elif act == 'w':
                self.shop.check_the_warehouse()
            elif act == 'p':
                self.shop.pr_compaign()
            elif act == 's':
                seller = self.shop.find_seller_by_name(input("Whose salary do you want to change (Write the seller's name)?"))
                if seller:
                    action = input(f"Do you want to raise or decrease the salary? (raise/decrease): ").lower()
                    percent = float(input("Enter the percentage to change the salary (number only)"))
                    if action == 'raise':
                        self.shop.raise_salary(seller, percent)
                    elif action == 'decrease':
                        self.shop.decrease_salary(seller, percent)
                    else:
                        print(f"You wrote a wrong name.")
            elif act == 'h':
                self.shop.hire_stuff()
            elif act == 'c':
                self.shop.show_stuff()
            elif act == 'd':
                self.shop.delete_instrument()
                self.shop.check_the_warehouse()
            elif act == 'e':
                break
            else:
                print('You chose a wrong action.')
                if input('Do you want to continue? (y/n)') == 'y':
                    continue
                else:
                    break
            

        



    def run(self):
        self.start_of_the_game()
        while True:
            print("Starting a new day...")
            self.working_day()
            self.end_of_day()
            if input("Want to start the next day? (y/n): ") == 'n':
                if input("Do you want to change something in your shop? (y/n)") == 'y':
                    self.between_days()
                    if input("Want to start the next day? (y/n): ") == 'n':
                        break
                    else:
                        continue
                else:
                    break
            else:
                continue
```
