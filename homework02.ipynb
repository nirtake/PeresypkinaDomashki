````python
import spacy
import en_core_web_sm
nlp = en_core_web_sm.load()

folder_path = input('Put in the path to the folder with files to compare with. ')

with open(f'{folder_path}\\news.txt', 'r', encoding='utf-8') as f:
    news = f.read()

with open(f'{folder_path}\\lovestory.txt', 'r', encoding='utf-8') as f:
    ls = f.read()

with open(f'{folder_path}\\medicin.txt', 'r', encoding='utf-8') as f:
    med = f.read()

with open(f'{folder_path}\\truecrime.txt', 'r', encoding='utf-8') as f:
    tc = f.read()

with open(f'{folder_path}\\sonnets.txt', 'r', encoding='utf-8') as f:
    sonnets = f.read()

with open(f'{folder_path}\\memorandum.txt', 'r', encoding='utf-8') as f:
    memo = f.read()

with open(f'{input("Put in path to the file")}', 'r', encoding='utf-8') as f:
    input_text = f.read()



def sense_words(file):
    doc = nlp(file)
    TL = []
    for token in doc:
        if token.pos_ == "ADJ" or token.pos_ == "ADV" or token.pos_ == "INTJ" or token.pos_ == "NOUN" or token.pos_ == "PROPN" or token.pos_ == "VERB" or token.pos_ == "SYM":
            TL.append(token.text.lower())
        else:
            continue
    return set(TL)

def similar_words(set1, set2):
    """set1 -- список слов из сравниваемого текста, set2 -- список слов из текста, с которым сравниваем"""
    SW = []
    for w in set1:
        if w in set2:
            SW.append(w)
    SW
    return len(set2)


news = similar_words(sense_words(input_text), sense_words(news))
ls = similar_words(sense_words(input_text), sense_words(ls))
med = similar_words(sense_words(input_text), sense_words(med))
tc = similar_words(sense_words(input_text), sense_words(tc))
sonnets = similar_words(sense_words(input_text), sense_words(sonnets))
memo = similar_words(sense_words(input_text), sense_words(memo))

DOCS = [news, ls, med, tc, sonnets, memo]

if max(DOCS) == news:
    print('Your texts resembles news.')
elif max(DOCS) == ls:
    print('Your text resembles a love story.')
elif max(DOCS) == med:
    print('Your text resembles a medicine text.')
elif max(DOCS) == tc:
    print('Your text resembles true crime.')
elif max(DOCS) == sonnets:
    print("Your text resembles a Shakespear's sonnet.")
elif max(DOCS) == memo:
    print('Your text resembles an official memorandum.')


#Я не смогла придумать как сделать так, чтобы не пришлось открывать все файлы из папки по отдельности и обрабатывать каждый, получилось как-то слишком длинно(
````

````python
import pyconll


text = pyconll.load_from_file(f'D:\\test.conllu')
c = 0
c2 = 0

for sentence in text:
    for token in sentence:
        if token.deprel == "dislocated":
            print(f'{sentence}\n{token.form}\t{token.deprel}')
            c += 1

if c == 0:
    print('No sentences with dislocated tag are found.')

for sentence in text:
    for token in sentence:
        if token.deprel == "discourse":
            print(f'{sentence.text}\n{token.form}\t{token.deprel}')
            c2 += 1

if c2 == 0:
    print('No sentences with discourse tag are found.')
````

Проблемы у него встречаются в определении части речи и приведении к лемме, при этом если часть речи определена правильно, то морфологические характеристики тоже определяются верно. Например, он посчитал, что глагол пр. вр. "grät" это существительное в единственном числе. А вот форму настоящего времени того же глагола, которая окончанием как раз может напоминать существительное ("gråter") он определил верно (правда, только один раз, а когда эта же форма встретилась в тексте ниже, он ее уже обозвал существительным во множественном числе, хотя она стояла прямо после подлежащего.) При этом с неправильными глаголами, у которых основы разных форм сильно друг от друга отличаются, он вполне справляется (är привел к инфинитиву vara, ничего себе (правда сразу после него совершенно регулярную форму прошедшего времени к совершенно стандартной форме инфинитива вообще без чередований привести не смог)). Еще проблемы возникали с притяжательными формами существительных (генетивами), которые вообще абсолютно регулярные у существительных (просто s на конце слова прибавляется). Форму "Daniels" он как генетив от Daniel не узнал, решил, что это новое существительное, хотя с формой генетива у местоимения справился ("hans" -- "han"), а у местоимений эта форма не так регулярно образовывается, это просто конкретно с этим повезло.
Мне показалось, что проблемы с глаголами возникают, если в предложении их больше одного и особенно если есть однородные сказуемые. В таких случаях он часто один глагол (причем необязательно первый) за человека (то есть за глагол) считал, а второй называл существительным, но не всегда.
В некоторых случаях, где форма слова напоминает другую часть речи, тоже не справлялся, например, причастие назвал существительным. Форму генетива местоимений упорно отказывается приводить к лемме, оставляет в той же форме.
Зато грамматические признаки для верно определенных частей речи назвал правильно, кажется, вообще везде (там, правда, для каждого признака по два варианта в парадигме, но хорошо же).
