```python
import pyconll


text = pyconll.load_from_file(f'D:\\ru_syntagrus-ud-train-b (1).conllu')

from typing_extensions import AsyncGenerator

class Predicativ: 
    def __init__(self, name):
        self.name = name
    
    def __str__(self):
        return f"{self.name}"
    
    def m_metr(self):
        count = 0
        for sentence in text:
            for token in sentence:
                if token.form == self.name():
                    x = token.id
                    for token in sentence:
                        if token.head == x and token.form == "мне":
                            count += 1
        return count
    
    def e_metr(self):
        count = 0
        for sentence in text:
            for token in sentence:
                if token.form == self.name:
                    x = token.id
                    for token in sentence:
                        if token.head == x and (token.form == "ему" or token.form == "ей"):
                            count += 1
        return count
    
    def m_e_metr(self):
        if e_metr(self.name) > 0:
            return m_metr(self.name) / e_metr(self.name)
        else:
            return 0

    def N_metr(self):
        count = 0
        for sentence in text:
            for token in sentence:
                if token.form == self.name:
                    count += 1
        return count
    
    def m_N_metr(self):
        if N_metr(self.name) > 0:
            return m_metr(self.name) / N_metr(self.name)
        else:
            return 0
        
    def summary(self):
        if m_metr(self) < 10:
            print(f'm-metric for this text is extremely low.')
        else:
            print(f'm-metric for this text is {m_metr(self)}')
        if e_metr(self) < 10:
            print(f'e-metric for this text is extremely low.')
        else:
            print(f'e-metric for this text is {e_metr(self)}')
        if m_e_metr(self) > 1.25:
            print('DPS predicatives in text are autoreferential')
        elif 1 <= m_e_metr(self) < 1.25:
            print('DPS predicatives in text are mildy non-autoreferential')
        else:
            print('DPS predicatives in text are strictly non-autoreferential')
        print(f'N-metric for this text is {N_metr(self)}')
        print(f'm/N-metric for this text is {m_N_metr(self)}')
        
        


#Когда я записала все отдельными функциями и передавала в них строки, все считалось (хотя цифры маленькие получались), в классах же на все выдает нули, хотя вроде бы те же функции туда вставила.
        
    

    
    

    




    
    
```
